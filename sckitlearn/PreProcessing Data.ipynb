```python
import pandas as pd
```

```python
# Lendo um arquivo de dados
exam_data = pd.read_csv('exams.csv')
```

```python
exam_data.head()
```

```python
exam_data.shape
```

```python
# Calculando a média de cada matéria 
math_average = exam_data['math score'].mean()
reading_average = exam_data['reading score'].mean()
writing_average = exam_data['writing score']. mean()

print(f' Math avg: {math_average} \n Reanding avg:{reading_average} \n Writing agv: {writing_average}')
```

```python
from sklearn import preprocessing

# O preprossecing usando scale do sklear padroniza os dados para uma leitura mais simples na ML
# Ele divide os dados pela media e subtrai pelo desvio padrão
exam_data[['math score']] = preprocessing.scale(exam_data[['math score']])
exam_data[['writing score']] = preprocessing.scale(exam_data[['writing score']])
exam_data[['reading score']] = preprocessing.scale(exam_data[['reading score']])
```

```python
exam_data.head()
```

```python
# Calculando as novas médias
math_average = exam_data['math score'].mean()
reading_average = exam_data['reading score'].mean()
writing_average = exam_data['writing score']. mean()

print(f' Math avg: {math_average} \n Reanding avg:{reading_average} \n Writing agv: {writing_average}')
```

```python
# Representação de dados categoricos
le = preprocessing.LabelEncoder()
exam_data['gender'] = le.fit_transform(exam_data['gender'].astype(str))
```

```python
# Perceba que esse preprocessamento binarizou a escala de genero, 0 = male, 1 = female
exam_data.gender.unique()
```
